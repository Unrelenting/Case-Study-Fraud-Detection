```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
df['is_new_user'] = (df['user_age'] == 0).astype(int)
df['venue_name_exist'] = (df['venue_name'].isnull()).astype(int)
fraud_list = ['fraudster', 'fraudster_att', 'fraudster_event']
df['fraud'] = (df['acct_type'].apply(lambda x: x in fraud_list)).astype(int)
df['payout_type_empty'] = (df['payout_type']=='').astype(int)
cols = ['payout_type_empty', 'venue_name_exist', 'is_new_user','sum_ticket_sold_iszero', 'channel_g1', 'channel_g2', 'num_payouts_above_30','channel_g3', 'delivery_method_iszero']
X = df[cols].values
y = df['fraud'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
lr = LogisticRegression()
model = lr.fit(X_train, y_train)
print "accuracy is {}".format(model.score(X_test, y_test))
print "f1 score is {}".format(f1_score(model.predict(X_test), y_test))
print "cross val f1 score is {}".format((np.mean(cross_val_score(lr, X_train, y_train, scoring='f1'))))
```

```python
df['num_payouts_above_30'] = (df['num_payouts']> 30).astype(int)
```

```python
df['delivery_method_iszero'] = (df['delivery_method'] == 0).astype(int)
```

```python
def sum_ticket_sold(lst):
    count = 0
    for l in lst:
        count += l['quantity_sold']
    return count    
df['sum_ticket_sold'] = df['ticket_types'].apply(lambda x: sum_ticket_sold(x))
```

```python
df['sum_ticket_sold_iszero'] = (df['sum_ticket_sold'] == 0).astype(int)
sum(df['sum_ticket_sold_iszero'])
```

```python
def convert_channel(row):
    if row == 0:
        return 0
    elif row == 8:
        return 8
    else:
        return 1
df['converted_channels'] = df['channels'].apply(lambda x: convert_channel(x))
```

```python
pd.get_dummies(df['converted_channels'], cols)
```

```python
from __future__ import division
c1 = Counter(df['channels'][df.fraud==1])
count = sum(c1.values())
for c in c1.keys():
    c1[c] /= count
c0 = Counter(df['channels'][df.fraud==0])
count = sum(c0.values())
for c in c0.keys():
    c0[c] /= count
```
